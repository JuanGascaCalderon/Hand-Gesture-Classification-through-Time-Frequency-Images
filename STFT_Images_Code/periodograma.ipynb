## Periodograma

```python
#Importación de Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
df = pd.read_excel(r"C:\Users\JUANPABLOGASCA\Desktop\PARCIAL_SISTEMAS_INTELIGENTES\P10.xlsx")
df
```

```python
# variables para iterar la segmentacion superpuesta
i = 0
j = 0
h = 0
k = 300
```

```python

#ciclo para generar las images para todas los canales y posturas
# Clase 1
i = 0
j = 0
fs = 200
for k in range(300, 6001, 150):  
    while j < 8:
        data1 = df.iloc[:, j]
        data2 = data1[i:k]  
        f, Pxx = signal.periodogram(data2, fs)
        plt.figure(figsize=(8, 4))  # ajusta el tamaño de la figura
        plt.plot(f, Pxx)
        plt.xlabel('Frecuencia (Hz)', fontsize=12)  # ajusta el tamaño del texto
        plt.ylabel('Potencia', fontsize=12)
        plt.tick_params(labelsize=10)  # ajusta el tamaño de las etiquetas del eje
        plt.tight_layout()  # ajusta el espacio entre los elementos del gráfico
        plt.savefig(rf'C:\Users\JUANPABLOGASCA\Desktop\PARCIAL_SISTEMAS_INTELIGENTES\DATASET\P10_PERIODOGRAMA\CLASE_1\{i}_{k}_{j}_PERIODOGRAMA.png', bbox_inches='tight', pad_inches=0)
        plt.clf()  # limpia la figura para la siguiente iteración
        
        j += 1
    j = 0
    i += 150
```
